Simple Quiz Application: Design a quiz app with multiple-choice questions. Track scores and give feedback on correct or incorrect answers using if-else statements and loops

```python
quiz = [
    {
        "question": "What is the output of print('Hello' + 'World')?",
        "choices": ["A. Hello World", "B. HelloWorld", "C. Hello+World", "D. Syntax Error"],
        "answer": "B"
    },
    {
        "question": "What is the output of the following code: print(2 ** 3)?",
        "choices": ["A.6", "B. 9", "C. 8", "D. 12"],
        "answer": "C"
    },
    {
        "question": "Which of the following data types is mutable?",
        "choices": ["A. int", "B. float", "C. tuple", "D. list"],
        "answer": "D"
    },
    {
        "question": "How do you start a comment in Python?",
        "choices": ["A. //", "B. /*", "C. #", "D. <!--->"],
        "answer": "C"
    },
    {
        "question": "What will be the output of the following code: print(10 // 3)?",
        "choices": ["A. 3", "B. 3.33", "C. 3.0", "D. Error"],
        "answer": "A"
    },
    {
        "question": "Which of the following statements is used to create a loop that iterates over a sequence in Python?",
        "choices": ["A. while", "B. for", "C. loop", "D. repeat"],
        "answer": "B"
    },
    {
         "question": "What does the len() function do in Python?",
        "choices": ["A. Returns the last element of a list", "B. Returns the length of a list or string", "C. Returns the largest number in a list", "D. Returns a random element from a list"],
        "answer": "B"
    }]
```

```python
score = 0

print("Welcome to the Quiz! Choose an Option A, B, C, or D.\n")

for i in quiz:
    print(i["question"])
    for choice in i['choices']:
        print(choice)
     
    user=input('Your answer: ').upper()
    if user == i["answer"]:
        print("Correct!\n")
        score += 1
    else:
        print(f"Incorrect. The correct answer is {i['answer']}.\n")


print(f"Quiz Complete! Your final score is: {score}/{len(quiz)}")
```
